```python
import pandas as pd
```

```python
Rocket_TEp5 = pd.read_csv('./results/Rocket/TEp5_cw.csv',index_col=0)
Rocket_TEp4 = pd.read_csv('./results/Rocket/TEp4_cw.csv',index_col=0)
Rocket_TEp3 = pd.read_csv('./results/Rocket/TEp3_cw.csv',index_col=0)
Rocket_TEp2 = pd.read_csv('./results/Rocket/TEp2_cw.csv',index_col=0)
Rocket_TEp1 = pd.read_csv('./results/Rocket/TEp1_cw.csv',index_col=0)
```

```python
Rocket_df = pd.concat([Rocket_TEp5,Rocket_TEp4,Rocket_TEp3,Rocket_TEp2,Rocket_TEp1])
Rocket_df = Rocket_df.reset_index(drop=True)
```

```python
CIF_TEp5 = pd.read_csv('./results/CIF/TEp5_cw.csv',index_col=0)
CIF_TEp4 = pd.read_csv('./results/CIF/TEp4_cw.csv',index_col=0)
CIF_TEp3 = pd.read_csv('./results/CIF/TEp3_cw.csv',index_col=0)
CIF_TEp2 = pd.read_csv('./results/CIF/TEp2_cw.csv',index_col=0)
CIF_TEp1 = pd.read_csv('./results/CIF/TEp1_cw.csv',index_col=0)
```

```python
CIF_df = pd.concat([CIF_TEp5,CIF_TEp4,CIF_TEp3,CIF_TEp2,CIF_TEp1])
CIF_df = CIF_df.reset_index(drop=True)
```

```python
Rocket_df['cw'] = ''
Rocket_df['Testing Partition'] = ''

for i in range(len(Rocket_df)):
    Rocket_df.loc[i,'cw'] = int(Rocket_df.loc[i,'Experiments'].split('_cw')[-1])
    Rocket_df.loc[i,'Testing Partition'] = int(Rocket_df.loc[i,'Experiments'].split('_cw')[0].split('TEp')[-1])
```

```python
CIF_df['cw'] = ''
CIF_df['Testing Partition'] = ''

for i in range(len(CIF_df)):
    CIF_df.loc[i,'cw'] = int(CIF_df.loc[i,'Experiments'].split('_cw')[-1])
    CIF_df.loc[i,'Testing Partition'] = int(CIF_df.loc[i,'Experiments'].split('_cw')[0].split('TEp')[-1])
```
