```python
result_exp1 = pd.read_csv('./results/Exp1/RF_Exp1_24features_cwALL_bootstrap10.csv')
result_exp1_100 = pd.read_csv('./results/Exp1/RF_Exp1_24features_cwALL_bootstrap100.csv')
result_exp1_1000 = pd.read_csv('./results/Exp1/RF_Exp1_24features_cwALL_bootstrap1000.csv')

result_exp2 = pd.read_csv('./results/Exp2/RF_Exp2_24features_cwALL_bootstrap10.csv')
result_exp2_100 = pd.read_csv('./results/Exp2/RF_Exp2_24features_cwALL_bootstrap100.csv')
result_exp2_1000 = pd.read_csv('./results/Exp2/RF_Exp2_24features_cwALL_bootstrap1000.csv')

result_exp3 = pd.read_csv('./results/Exp3/RF_Exp3_24features_cwALL_bootstrap10.csv')
result_exp3_100 = pd.read_csv('./results/Exp3/RF_Exp3_24features_cwALL_bootstrap100.csv')
result_exp3_1000 = pd.read_csv('./results/Exp3/RF_Exp3_24features_cwALL_bootstrap1000.csv')
```

```python
def series_df(df):
    window_size = []
    class_weight = []
    for i in range(len(df)):
        window_size.append(df.loc[i,'Experiment'].split('TEp4_')[-1].split('_cw')[0])
        class_weight.append(int(df.loc[i,'Experiment'].split('_cw')[-1].split('_bin')[0]))
    df['window_size'] = window_size
    df['class_weight'] = class_weight
        
    TSS = []
    HSS = []
    label = []
    for ws in df['window_size'].unique():
        for cw in df['class_weight'].unique():
            sub_df = df.loc[(df['class_weight'] == cw) & (df['window_size'] == ws)]
            TSS.append(sub_df['TSS'].values)
            HSS.append(sub_df['HSS'].values)
            label.append(ws+'_cw:{1:'+str(cw)+'}')
            
    return TSS, HSS, label
```

```python
bs10_exp1_TSS, bs10_exp1_HSS, bs10_exp1_label = series_df(result_exp1)
bs100_exp1_TSS, bs100_exp1_HSS, bs100_exp1_label = series_df(result_exp1_100)
bs1000_exp1_TSS, bs1000_exp1_HSS, bs1000_exp1_label = series_df(result_exp1_1000)

bs10_exp2_TSS, bs10_exp2_HSS, bs10_exp2_label = series_df(result_exp2)
bs100_exp2_TSS, bs100_exp2_HSS, bs100_exp2_label = series_df(result_exp2_100)
bs1000_exp2_TSS, bs1000_exp2_HSS, bs1000_exp2_label = series_df(result_exp2_1000)

bs10_exp3_TSS, bs10_exp3_HSS, bs10_exp3_label = series_df(result_exp3)
bs100_exp3_TSS, bs100_exp3_HSS, bs100_exp3_label = series_df(result_exp3_100)
bs1000_exp3_TSS, bs1000_exp3_HSS, bs1000_exp3_label = series_df(result_exp3_1000)
```

```python
fig, axs = plt.subplots(2, figsize=(12, 7))
axs[0].boxplot(bs10_exp1_TSS)
axs[0].set(ylabel='TSS')
axs[1].boxplot(bs10_exp1_HSS)
axs[1].set(ylabel='HSS')
axs[1].set_xticklabels(bs10_exp1_label, fontsize=10, rotation=75)

for ax in axs.flat:
    ax.label_outer()
plt.savefig('./plots/Bootstrap10_Exp1_boxplot.png', bbox_inches='tight')
```
